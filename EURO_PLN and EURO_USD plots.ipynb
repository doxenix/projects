# Storytelling Data Visualization on Exchange Rates

In this project we will take a look at data set with Daily Exchange Rates per Euro 1999-2021. We can find here comparison to most of world values.

We will create 2 plots using matplotlib library, one with currency fluctuations during financial crisis and COVID-19 in EURO/USD and second one in EURO/PLN.

Data set was made by Daria Chemkaeva and is available [here](https://www.kaggle.com/lsind18/euro-exchange-daily-rates-19992020)

Let's start wiith opening our data set.

```python
#opening file
import pandas as pd

exchange_rates = pd.read_csv('euro-daily-hist_1999_2020.csv')
#display 5 first rows
exchange_rates.head()
```

Our data set contains 41 columns, 40 diffrent currencies included.

```python
#display 5 last rows
exchange_rates.tail()
```

```python
#show info about dataframe columns
exchange_rates.info()
```

Our dataset contains 41 columns with diffrent currencies. Some of them have null values. We also see that most data type of each column are objects, only few are numeric.

```python
#rename colums 
exchange_rates.rename(columns={'[US dollar ]': 'US_dollar',
                               'Period\\Unit:': 'Time',
                                '[Polish zloty ]' : 'PLN'
                              },
                      inplace=True)

#convert Time column to datatime type
exchange_rates['Time'] = pd.to_datetime(exchange_rates['Time'])
#sorting by time
exchange_rates.sort_values('Time',inplace=True)
```

```python
#isolate Time and USD
euro_to_dollar = exchange_rates[['Time', 'US_dollar']]
#display 5 first rows
euro_to_dollar.head()
```

```python
#isolate Time and PLN
euro_to_pln = exchange_rates[['Time', 'PLN']]
#display 5 first rows
euro_to_pln.head()
```

```python
euro_to_pln['PLN'].value_counts().sort_values()
```

```python
euro_to_dollar['US_dollar'].value_counts().sort_values()
```

62 values in US_dollar and PLN column are '-' character. We will remove them

```python
#removing rows with '-' in US_dollar column
euro_to_dollar = euro_to_dollar[euro_to_dollar['US_dollar'] != '-']
#convert to float
euro_to_dollar['US_dollar'] = euro_to_dollar['US_dollar'].astype(float)

#removing rows with '-' in PLN column
euro_to_pln = euro_to_pln[euro_to_pln['PLN'] != '-']
#convert to float
euro_to_pln['PLN'] = euro_to_pln['PLN'].astype(float)
```

```python
#import matplotlib
import matplotlib.pyplot as plt
#add this row to enables Jupyter to generate the graphs
%matplotlib inline
plt.plot(euro_to_dollar['Time'],
         euro_to_dollar['US_dollar'])
plt.show()
```

```python
plt.plot(euro_to_pln['Time'],
         euro_to_pln['PLN'])
plt.show()
```

```python
#creat new column with moving average 30 days USD
euro_to_dollar['rolling_mean'] = euro_to_dollar['US_dollar'].rolling(30).mean()
#display 5 last rows
euro_to_dollar.tail()
```

```python
#creat new column with moving average 30 days PLN
euro_to_pln['rolling_mean'] = euro_to_pln['PLN'].rolling(30).mean()
#display 5 last rows
euro_to_pln.tail()
```

```python
#isolate rows with year 2006-2021 only USD
euro_to_dollar_06_21 = euro_to_dollar.copy()[(euro_to_dollar['Time'].dt.year >= 2006)]
#display 5 last rows
euro_to_dollar_06_21.tail()
```

```python
euro_to_dollar_18_21 = euro_to_dollar.copy()[euro_to_dollar['Time'].dt.year >= 2018]
```

```python
#isolate rows with year 2007-2009 only USD
euro_to_dollar_07_09 = euro_to_dollar.copy()[(euro_to_dollar['Time'].dt.year >= 2007) & (euro_to_dollar['Time'].dt.year <= 2009)]
euro_to_dollar_07_09.head()
```

```python
#isolate rows with year 2006-2021 only PLN
euro_to_pln_06_21 = euro_to_pln.copy()[(euro_to_pln['Time'].dt.year >= 2006)]
#display 5 last rows
euro_to_pln_06_21.tail()
```

```python
#isolate rows with year 2007-2009 only PLN
euro_to_pln_18_21 = euro_to_pln.copy()[euro_to_pln['Time'].dt.year >= 2018]

#isolate rows with year 2007-2009 only PLN
euro_to_pln_07_09 = euro_to_pln.copy()[(euro_to_pln['Time'].dt.year >= 2007) & (euro_to_pln['Time'].dt.year <= 2009)]
euro_to_pln_07_09.head()
```

## Creating EURO-USD plot

```python
#import matplotlib style and dates, import datetime
import matplotlib.style as style
from datetime import datetime
from matplotlib.dates import date2num


### Adding the FiveThirtyEight style
style.use('fivethirtyeight')

### Adding the plot
fig, ax = plt.subplots(figsize=(10,4))
ax.plot(euro_to_dollar_06_21['Time'], euro_to_dollar_06_21['rolling_mean'], 
       linewidth=1, color='#A6D785')

### Highlighting the 2007-2008 period
ax.plot(euro_to_dollar_07_09 ['Time'],
        euro_to_dollar_07_09 ['rolling_mean'],
        linewidth=3, color='#e23d28')

### Highlighting the 2018-2021 period
ax.plot(euro_to_dollar_18_21 ['Time'],
        euro_to_dollar_18_21 ['rolling_mean'],
        linewidth=3, color='#e23d28')

### Highlihting the peak of the crisis
ax.axvspan(date2num(datetime(2007,1,1)), date2num(datetime(2009,1,1)), 
           label="2009 Recession",color="gray", alpha=0.3)
ax.axvspan(date2num(datetime(2020,1,1)), date2num(datetime(2021,1,8)),
           label="2009 Recession",color="gray", alpha=0.3)

# Set x-labels            
ax.set_xticks([date2num(datetime(2007,1,1)), date2num(datetime(2009,1,1)), date2num(datetime(2011,1,1)), 
               date2num(datetime(2013,1,1)), date2num(datetime(2015,1,1)), date2num(datetime(2017,1,1)), 
              date2num(datetime(2019,1,1)), date2num(datetime(2021,1,1))])
# Set y-labels
ax.set_yticks([1.0, 1.1, 1.2, 1.3, 1.4, 1.5, 1.6])
ax.set_yticklabels(['', '1.1', '', '1.3', '', '1.5', ''])

# Adding text under 

### Adding title Highlihting peak
ax.text(date2num(datetime(2006,12,15)), 1.60, "Financial crisis", color = 'gray',
       size=11, weight='bold')
ax.text(date2num(datetime(2019,11,1)), 1.60, "COVID-19", color = 'gray',
       size=11, weight='bold')

ax.text(date2num(datetime(2006,1,1)), 1.7, "EURO-USD fluctations during financial crisis and COVID-19",
       weight='bold', size=15)
ax.text(date2num(datetime(2006,1,1)), 1.66, "Euro-USD exchange rates 2006 - 2010 and 2016 - 2021",
        size=12)

### Add some transparency to the grid
ax.grid(alpha=0.5)

ax.text(date2num(datetime(2006,1,1)), 0.9, 'Â©DominikZalewski' + ' '*127 + 'Source: European Central Bank',
        color = '#f0f0f0', backgroundcolor = '#4d4d4d',
        size=10)

plt.show()
```
